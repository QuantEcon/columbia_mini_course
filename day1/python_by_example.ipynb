### While loops

```python
ts_length = 100
系_values = []
i = 0

while i < ts_length:
    e = np.random.randn()
    系_values.append(e)
    i = i + 1
    
plt.plot(系_values, 'b-')
plt.show()
```

### Functions

```python
def f(x):
    return 2 * x
```

```python
f(3)
```

```python
def g(x, y):
    return 2 * max(x, y)
```

```python
g(3, 4)
```

With explicit branching:

```python
def g2(x, y):
    """
    Does the same thing as g
    
    """
    if x > y:
        mx = x
    else:
        mx = y
        
    return 2 * mx
        
```

```python
g2(3, 4)
```

Let's write a function for populating a Python list with draws from a standard normal distribution:

```python
def generate_data(n):
    系_values = []
    for i in range(n):
        e = np.random.randn()
        系_values.append(e)
    return 系_values
```

```python
data = generate_data(100)
```

```python
data[0:4]
```

```python
plt.plot(data, 'b-')
plt.show()
```

A fancier function:

```python
def generate_data(n, generator_type):
    系_values = []
    for i in range(n):
        if generator_type == 'U':
            e = np.random.uniform(0, 1)
        else:
            e = np.random.randn()
        系_values.append(e)
    return 系_values
```

```python
data = generate_data(100, 'U')
```

```python
data[0:4]
```

```python
plt.plot(data, 'b-')
plt.show()
```

Here's a nicer function that does the same thing:

```python
def generate_data(n, generator_type):
    系_values = []
    for i in range(n):
        e = generator_type()
        系_values.append(e)
    return 系_values
```

Note: we're passing a function to a function

```python
data = generate_data(100, np.random.uniform)
plt.plot(data, 'b-')
plt.show()
```

### List comprehensions

List comprehensions are another interface to loops

Used to create new lists from old:

```python
animals = ['dog', 'cat', 'bird']
plurals = [animal + 's' for animal in animals]
plurals
```

```python
doubles = [2 * x for x in range(8)]
doubles
```

```python
x = [np.random.randn() for i in range(100)]
plt.plot(x, 'b-', label="white noise")
plt.legend()
plt.show()
```
